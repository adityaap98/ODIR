# **Code running instructions**
## Ocular Disease Intelligent Recognition

Kindly run the Jupyter notebook containing the code on Google Colaboratory platform as it offers free GPU and a lot of pre-installed libraries needed for machine learning problems.

This Jupyter notebook will serve as a step-by-step guide to run the [main code](https://colab.research.google.com/drive/1oVh53jWPXuqGPL71myaGJyipydQ-d56k) notebook.

---



## Step 1: Downloading the "*kaggle.json*" file
This file is needed to directly load the ODIR dataset (from Kaggle) onto the Colaboratory. Download it from [here](https://drive.google.com/file/d/1CH1HjBiy3o6hjKQZshxtQdrfCRks9ngm/view?usp=sharing).


## Step 2: Uploading the "*kaggle.json*" file on Colaboratory
*  Now go to the [main code](https://colab.research.google.com/drive/1oVh53jWPXuqGPL71myaGJyipydQ-d56k) notebook.
*  On the left side of the notebook, find and click on an option named "*files*" (as shown below).

download.png

*  This will let the notebook to automatically connect to the server. **Wait until the "*sample_data*" folder is displayed**. Then, right click the empty space below "*sample_data*" and there you will find the option to upload a file. Click upload, and select *kaggle.json* from local downloads (see the screenshots below)

download (1).png

kaggle.PNG

## Step 3: Running the Code

To run the Jupyter notebook, one can either run each cell one after the other, or run all the cells at once by pressing *Ctrl + F9*.

> ***Note: You will be asked to choose a model out of the four models. Please select a model of your choice!***

> Next, if you wish to run another model, just run the cells below the heading '*Calling the model*' again.
